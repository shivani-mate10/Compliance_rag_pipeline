```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import OllamaEmbeddings

# Initialize the same embedding model
embedding_model = OllamaEmbeddings(model="nomic-embed-text")

# Load the existing vector stores from disk
db = Chroma(
    persist_directory="dbv1/chroma_db",
    embedding_function=embedding_model
)

annual_db = Chroma(
    persist_directory="dbv3/chroma_db", 
    embedding_function=embedding_model
)

# Now you can query them
rule_query = "what is net profit calculation rule"
rule_chunks = db.similarity_search(rule_query, k=2)

report_query = "what is net profit of company ABC"
report_chunks = annual_db.similarity_search(report_query, k=2)

# View results
print("Rule chunks:", rule_chunks[0].page_content)
print("\nReport chunks:", report_chunks[0].page_content)
```

```python
from langchain_core.messages import HumanMessage

#from langchain.schema import HumanMessage
from langchain_ollama import ChatOllama

# Initialize your LLM
llm = ChatOllama(model="llama3.2:1b", temperature=0)

# Combine rule chunks (top 2)
rule_text = "\n\n---\n\n".join([chunk.page_content for chunk in rule_chunks])

# Combine report chunks (top 2)  
report_text = "\n\n---\n\n".join([chunk.page_content for chunk in report_chunks])

llm_prompt = f"""
You are an AI compliance checker.

Compliance Rules:
{rule_text}

Annual Report Data:
{report_text}

Check if the annual report follows the rule of net profit. 
Explain your reasoning step by step and indicate PASS or FAIL.
"""

message = HumanMessage(content=llm_prompt)

# Get response from LLM
response = llm.invoke([message])
print(response.content)
```
